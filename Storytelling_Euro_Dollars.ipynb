```python
import pandas as pd

exchange_rate = pd.read_csv("euro-daily-hist_1999_2020.csv")
```

```python
exchange_rate.info()
```

```python
print("\\U4534")
```

```python
exchange_rate.rename(columns = {"[US dollar ]":"US_dollar","Period\\Unit:":"Time"},inplace = True)
```

```python
exchange_rate["Time"] = pd.to_datetime(exchange_rate["Time"])
```

```python
exchange_rate.sort_values("Time", inplace = True)
```

```python
exchange_rate.reset_index(drop = True, inplace = True)
```

```python
exchange_rate.head()
```

```python
euro_to_dollar = exchange_rate[["Time","US_dollar"]].copy()
euro_to_dollar["US_dollar"].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar.US_dollar != "-"]
```

```python
euro_to_dollar["US_dollar"] = euro_to_dollar["US_dollar"].astype(float)
```

```python
euro_to_dollar["US_dollar"].rolling(7).mean().head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"])
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)

plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"])
plt.title("Orginal Values", weight = "bold")


for i, rolling_mean in zip(range(2,7), [7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar["Time"], 
             euro_to_dollar["US_dollar"].rolling(rolling_mean).mean())
    plt.title("Rolling Window:" + str(rolling_mean), weight = "bold")

plt.tight_layout()
```

```python
euro_to_dollar["rolling_mean"] = euro_to_dollar["US_dollar"].rolling(30).mean()
```

```python
# Financial Crisis
```

```python
financial_crisis = \
euro_to_dollar[(euro_to_dollar["Time"].dt.year >= 2006) & (euro_to_dollar["Time"].dt.year <= 2009)]

financial_crisis_7_8 = \
euro_to_dollar[(euro_to_dollar["Time"].dt.year >= 2007) & (euro_to_dollar["Time"].dt.year <= 2008)]

financial_crisis_7_8["US_dollar"].max()
```
